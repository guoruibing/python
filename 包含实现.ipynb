```python
import xlrd
import numpy as np
import pandas as pd
#路径前加 r，读取的文件路径
file_path = r'C:\\Users\\Administrator\\Desktop\\companytype.xlsx'
#获取数据
data = xlrd.open_workbook(file_path)
#获取sheet
table = data.sheet_by_name('Sheet')
#获取总行数
nrows = table.nrows
#获取总列数
ncols = table.ncols
#获取第一列的数值
companytypecol_values = table.col_values(0)
pdcompanytypecol = pd.DataFrame(companytypecol_values)
pdcompanytypecol.columns = ['company']
pdcompanytypecol
```
